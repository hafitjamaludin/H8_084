## Line Plot
**Question** : Bagaimanakah Jumlah kasus di City of London dari tahun 2008-2016?


```python
data_tahuncl = pd.DataFrame(
               data[data["Wilayah"]== "City of London"].groupby(["Tahun"])["Value"].sum()
               ).sort_values(by='Tahun', ascending=True)
data_tahuncl
```

```python
data_tahuncl.plot(kind='line', figsize=(10,6))

plt.title('Total value City of London dari tahun 2008-2016')
plt.ylabel('Value')
plt.xlabel('Year')

plt.show()
```

**Analysis :** Dari plot diatas, Terlihat bahwa dari tahun 2008-2009 total kasus adalah 0, lalu terjadi peningkatan jumlah kasus pada tahun 2011-2012, dan kembali menurun pada tahun 20013. Pada tahun 2014-2016 mengalami kenaikan jumlah kasus.

## Histogram
**Question 1 :** Bagaimana jumlah kasus pada Westminster dari tahun 2008-2016?

```python
data_hist = pd.DataFrame(
            data[data["Wilayah"]== "Westminster"].groupby(["Tahun"])["Value"].sum()
            ).sort_values(by='Tahun', ascending=True)
data_hist.reset_index(inplace=True)
data_hist
```

```python
data_croydon = []
for i in range(len(data_hist)):
    for j in range(int(data_hist.Value[i])):
        data_croydon.append(data_hist.Tahun[i])
```

```python
plt.hist(data_croydon,bins=6, color='#b0db43')

plt.title('Histogram Total value Westminster tahun 2008-2016 ')
plt.show()
```

**Analysis :** Pada plot diatas, terlihat bahwa jumlah kasus di Westminster pada tahun 2008 cukup tinggi. Sedangkan pada tahun 2009-2010 mengalami penurunan. Sedangkan pada tahun 2011-2013 mengalami kenaikan jumlah kasus kemudian mengalami penrunan kembali di tahun 2014-2015. Pada tahun 2016 mengalami kenaikan kembali.

## Area Plot

**Question :** Jumlah Total value Wstrminster,Lambeth, dan Southwark pada tahun 2016 

```python
data['Tahun'] = data['Tahun'].astype('str')
```

```python
# memilih data wilayah westminster di tahun 2016
data_area1 = data[(data['Tahun']=='2016') & (data['Wilayah']=='Westminster')]

# mengelompokkan data dengan month 
data_area1 = (data_area1.groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area1 = pd.DataFrame(data_area1)
data_area1.head(12)
```

```python
# memilih data wilayah westminster di tahun 2016
data_area2 = data[(data['Tahun']=='2016') & (data['Wilayah']=='Lambeth')]

# mengelompokkan data dengan
data_area2  = (data_area2 .groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area2  = pd.DataFrame(data_area2 )
data_area2 .head(12)
```

```python
# memilih data wilayah Southwark di tahun 2016
data_area3  = data[(data['Tahun']=='2016') & (data['Wilayah']=='Southwark')]

# mengelompokkan data dengan
data_area3 = (data_area3.groupby('Bulan')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Bulan', ascending=True)
                 .head(12))
data_area3 = pd.DataFrame(data_area3)
data_area3.head(12)
```

```python
merged_df = pd.merge(data_area1, data_area2, on="Bulan")
df = pd.merge(merged_df, data_area3, on="Bulan")
df = df.rename(columns={'Value_x': 'Westminster', 'Value_y': 'Lambeth', 'Value': 'Southwark'}) 
df.set_index('Bulan')
```

```python
df.plot(kind='area', 
        color=['#b4da55','#fcf9f4','#f42069'] , #,'coolwarm_r', #'tab20_r',
        stacked=False,
        alpha =.8,
        x='Bulan',
        figsize=(12,6)# pass a tuple (x,y) size
        )
             
plt.title("Jumlah Total value Wstrminster,Lambeth, dan Southwark pada tahun 2016", fontsize=14)
plt.ylabel("Value")
plt.xlabel("Bulan")

plt.show()
```

**Analysis:** Pada Plot diatas, terlihat bahwa pada tahun 2016 wilayah Westminster terjadi kenaikan pada bulan ke-12. Pada wilayah lambeth terjadi kenaikan pada bulan ke-7. dan pada wilayah Southwark terjadi kenaikan pada bulan ke-10. 

## Pie Chart

**Question1 :** Bagiamanakah Jumlah value dari tahun 2008-2016?

```python
data_tahun = (data.groupby('Tahun')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Tahun', ascending=False)
                 .head(12))
data_tahun   = pd.DataFrame(data_tahun )
data_tahun
```

```python
colors = ["#7a0037","#bd457b", "#ff8abf", "#ffc2de", "#fcf9f4" ,"#f0fcda" ,"#e4febf", "#b4da55", "#7e8e01" ]
plt.figure(figsize=(8,6)) 

plt.pie(
    data_tahun['Value'],
    shadow=False,
    colors=colors,
    explode=(0.15, 0, 0, 0,0, 0 , 0,0 , 0),
    startangle=90,
    autopct='%1.1f%%',
    )


plt.axis('equal')
plt.legend(labels=data_tahun['Tahun'], loc="center left")


# View the plot
plt.title('Pie Chart Total Value Pertahun')
plt.tight_layout()
plt.show()
```

**Analysis :** Pada plot diatas, terlihat bahwa tahun tertinggi yaitu tahun 2012 dengan 12.0%, lalu tahun 2011 dengan 11.4%, lalu yang ketiga yaitu 2016 dengan 11.2%. Sedangkan presentasi terendah yaitu pada tahun 2014.

**Question 2 :** Bagaiamanakah Total kejahatan Mayor dari tahun tahun 2008-2016?

```python
data_mayor = (data.groupby('Kejahatan Mayor')['Value']
                 .sum()
                 .reset_index()
                 .sort_values(by='Kejahatan Mayor', ascending=False)
                 .head(12))
data_mayor   = pd.DataFrame(data_mayor)
data_mayor
```

```python
colors = ["#7a0037","#bd457b", "#ff8abf", "#ffc2de", "#fcf9f4" ,"#f0fcda" ,"#e4febf", "#b4da55", "#7e8e01" ]
plt.figure(figsize=(10,6)) 

plt.pie(
    data_mayor['Value'],
    shadow=False,
    colors=colors,
    explode=(0.15, 0, 0, 0,0, 0 , 0,0 , 0),
    startangle=90,
    autopct='%1.1f%%',
    pctdistance=1.2,
    )


plt.axis('equal')
plt.legend(labels=data_mayor['Kejahatan Mayor'], loc="lower left")


# View the plot
plt.title('Pie Chart Total Value Kejahatan Mayor')
plt.tight_layout()
plt.show()
```

**Analysis:** Pada plot diatas, terlihat bahwa Kategori major yang tertinggi yaitu Theft and Handling dengan 56.2%, lalu menyusul Violence Against the Person dengan 17.6%. dan yang ketiga yaitu Burgalary dengan 6.9%.
